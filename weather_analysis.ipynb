# 4. Rainfall Distribution:
Illustrate the distribution of rainfall through a histogram or kernel density
plot, highlighting common levels and outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('weather.csv')
rainfall_data = np.array(df['Rainfall'].values)

# Create a figure with subplots for histogram and KDE plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Histogram plot
sns.histplot(rainfall_data, ax=ax1, kde=False, bins=30, color='skyblue')
ax1.set_title('Rainfall Distribution (Histogram)')
ax1.set_xlabel('Rainfall Level')
ax1.set_ylabel('Frequency')

# KDE plot
sns.histplot(rainfall_data, ax=ax2, kde=True, bins=3, color='salmon')
ax2.set_title('Rainfall Distribution (KDE Plot)')
ax2.set_xlabel('Rainfall Level')
ax2.set_ylabel('Density')

# Highlight outliers (optional)
outlier_threshold = 80  # Define your threshold for outliers
outliers = rainfall_data[rainfall_data > outlier_threshold]
ax1.axvline(outlier_threshold, color='red', linestyle='--', label=f'Outlier Threshold ({outlier_threshold})')
ax2.axvline(outlier_threshold, color='red', linestyle='--', label=f'Outlier Threshold ({outlier_threshold})')

# Show common levels (optional)
common_levels = [40, 50, 60]  # Define your common levels
for level in common_levels:
    ax1.axvline(level, color='green', linestyle=':', label=f'Common Level ({level})')
    ax2.axvline(level, color='green', linestyle=':', label=f'Common Level ({level})')

# Show legend
ax1.legend()
ax2.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

#### It's clear that the rainfall on the fourth day is an outlier value and must be removed.

# 5. Seasonal Analysis:
Analyze average values of weather variables across different seasons and
visualize seasonal patterns with bar graphs.

```python
def get_season(row):
    min_temp = row['MinTemp']
    max_temp = row['MaxTemp']
    
    if (max_temp >= 35) or (min_temp >= 30):
        return 'Summer'
    elif (max_temp >= 25) and (min_temp >= 18):
        return 'Spring'
    elif (max_temp >= 15) and (min_temp >= 5):
        return 'Autumn'
    else:
        return 'Winter'
```

```python
# Apply the function to create the 'season' column
data['season'] = data.apply(get_season, axis=1)
data
```

```python
winter = data.loc[data["season"]=="Winter"]
summer = data.loc[data["season"]=="Summer"]
autumn = data.loc[data["season"]=="Autumn"]
spring = data.loc[data["season"]=="Spring"]

plt.plot(winter.iloc[:,0].values,label="Winter")
plt.plot(summer.iloc[:,0].values,label="Summer")
plt.plot(autumn.iloc[:,0].values,label="Autumn")
plt.plot(spring.iloc[:,0].values,label="Spring")
plt.xlabel("Day")
plt.ylabel("Minimum Temperature")
plt.legend()
```
